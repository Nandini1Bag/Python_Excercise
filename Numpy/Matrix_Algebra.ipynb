### Addition  and Scalar Multiplication
#### â€¢ The sum of two ğ‘š Ã— ğ‘› matrices ğ´ and ğµ is the ğ‘š Ã— ğ‘› matrix ğ´ + ğµ in which each element is the sum of the corresponding elements of ğ´ and ğµ.
#### â€¢ The product of a scalar ğ‘˜ and a matrix ğ´ is the matrix ğ‘˜ğ´, in which each element is ğ‘˜ times the corresponding element of ğ´.

```python
import numpy as np
```

```python
# Example 1: Matrix Addition & Subtraction using numpy
# Given matrices
A = np.array([[1, -8, 3],
              [9, 11, 2],
              [5, -4, 0]])

B = np.array([[-3, 0, 2],
              [12, 7, -1],
              [6, 8, -5]])

C = np.array([[41, -6, 11],
              [-9, -14, 2],
              [0, 10, 4]])

# Find: 2A + B - C
result = 2 * A + B - C

print("2A + B - C =")
print(result)
```

### MATRIX MULTIPLICATION
#### â€¢ The product ğ´ğµ of two matrices ğ´ and ğµ can be found only if the number of columns of ğ´ equals the number of rows of ğµ.
#### â€¢ If ğ´ is an ğ‘š Ã— ğ‘› matrix and ğµ is an ğ‘› Ã— ğ‘˜ matrix, the product ğ´ğµ will be an ğ‘š Ã— ğ‘˜ matrix
#### â€¢ If ğ´ is an ğ‘š Ã— ğ‘› matrix and ğµ is an ğ‘› Ã— ğ‘š matrix, the products ğ´ğµ and ğµğ´ can both be found. However, ğ´ğµ â‰  ğµğ´ in general
#### Example :-Matrix A (2Ã—3)     Matrix B (3Ã—2) 
#### Multiply them: First row, first column:(W)
####                First row, second column:(X)
####                Second row, first column:(Y)
####                Second row, second column:(Z)
#### Resulting Matrix C (2Ã—2):[[W,X],[Y,Z]]

```python
# Matrices
A = np.array([[1, -3],
              [7, 2],
              [-2, 5]])

B = np.array([[1, 0, -1],
              [3, 1, 4]])

# AB â†’ (3x2) Ã— (2x3) = (3x3)
AB = np.dot(A, B)
#behind the scene:-  C=AB
#First row of C:
#ğ‘11=(1)(1)+(âˆ’3)(3)=1âˆ’9=âˆ’8
#ğ‘12=(1)(0)+(âˆ’3)(1)=0âˆ’3=âˆ’3
#c13 =(1)(âˆ’1)+(âˆ’3)(4)=âˆ’1âˆ’12=âˆ’13
#Second row of C:
#c21 =(7)(1)+(2)(3)=7+6=13
#c22 =(7)(0)+(2)(1)=0+2=2
#c23 =(7)(âˆ’1)+(2)(4)=âˆ’7+8=1
#Third row of C:
#c31 =(âˆ’2)(1)+(5)(3)=âˆ’2+15=13
#c32 =(âˆ’2)(0)+(5)(1)=0+5=5
#c33 =(âˆ’2)(âˆ’1)+(5)(4)=2+20=22

# BA â†’ (2x3) Ã— (3x2) = Not Possible
try:
    BA = np.dot(B, A)
except ValueError as e:
    BA = "Multiplication not possible: " + str(e)

print("AB =")
print(AB)

print("\nBA =")
print(BA)
```
