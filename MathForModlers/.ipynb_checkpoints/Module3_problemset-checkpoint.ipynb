#### 1)graph x-y>=-3 , 2x+6y<=12  

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(-10, 10, 400)

# Boundary lines
y1 = x + 3                    # from x - y ≥ -3 → y ≤ x + 3
y2 = (-1/3) * x + 2           # from 2x + 6y ≤ 12 → y ≤ -1/3x + 2

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the boundary lines
plt.plot(x, y1, label='y = x + 3', color='blue')
plt.plot(x, y2, label='y = -1/3x + 2', color='green')

# Fill the feasible region
# Create a grid of points
X, Y = np.meshgrid(x, np.linspace(-10, 10, 400))

# Define the inequalities
ineq1 = Y <= X + 3
ineq2 = Y <= (-1/3) * X + 2

# Region that satisfies both
region = ineq1 & ineq2

# Plot the region
plt.contourf(X, Y, region, levels=[0.5, 1], colors=['lightgray'])

# Axis and grid
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the system of inequalities')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2)Caitlyn sells cell phone sweaters and keychains at the county fair. She wants to sell at least 10 crafts in total. She sells the cell phone 
#### sweaters for (dollar)2.00 and the keychains for (dollar)1.00. She needs to sell at least $13 worth of crafts in order to earn a profit.
#### a) Choose the system of inequalities that matches the given problem (where  is the number of cell phone sweaters and  is the number of keychains).
#### b) Will she make a profit if she sells 3 cell phone sweaters and 4 keychains?

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(0, 15, 400)

# Define the boundary lines
y1 = -x + 10       # from x + y >= 10
y2 = -2*x + 13     # from 2x + y >= 13

# Create a grid of points
X, Y = np.meshgrid(x, np.linspace(0, 15, 400))

# Inequality conditions
ineq1 = X + Y >= 10
ineq2 = 2*X + Y >= 13

# Feasible region (where both inequalities are true)
region = ineq1 & ineq2

# Plot the boundary lines
plt.plot(x, y1, label='x + y = 10', color='blue')
plt.plot(x, y2, label='2x + y = 13', color='green')

# Fill the feasible region
plt.contourf(X, Y, region, levels=[0.5, 1], colors=['lightgray'])

# Customize the plot
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.xlabel('x (Cell phone sweaters)')
plt.ylabel('y (Keychains)')
plt.title('System of Inequalities: Sales at the Fair')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

#### 3)Miranda runs a restaurant that sells two kinds of desserts. Miranda knows the restaurant must make at least 8 and at most 28 batches of the Mint #### Breezes. The restaurant must also make at least 17 and at most 60 batches of the Caramel Twists. Each batch of Mint Breezes takes 5 minutes in the #### oven, while each batch of Caramel Twists requires 4 minutes in the oven. The restaurant only has 320 minutes in the oven available. Suppose each   #### batch of Mint Breezes generates (dollar)2.91 in revenue, each batch of Caramel Twists generates (dollar)1.31, and Miranda needs to maximize revenue.
#### Let x be the number of batches of Mint Breezes and let y be the number of batches of Caramel Twists. Set up the linear programming model.

```python
from scipy.optimize import linprog

# Objective function coefficients (negated for maximization)
# Revenue: Z = 2.91x + 1.31y ⇒ maximize Z
# We negate because linprog minimizes by default
c = [-2.91, -1.31]

# Inequality constraint matrix (for oven time): 5x + 4y ≤ 320
A = [[5, 4]]
b = [320]

# Variable bounds
# x (Mint Breezes): between 8 and 28
# y (Caramel Twists): between 17 and 60
x_bounds = (8, 28)
y_bounds = (17, 60)

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Check if the optimization was successful
if result.success:
    # Extract values
    x_opt = result.x[0]
    y_opt = result.x[1]
    max_revenue = -result.fun  # negate back to get maximum

    print("✅ Optimal Solution Found:")
    print(f"Mint Breezes (x): {x_opt:.0f}")
    print(f"Caramel Twists (y): {y_opt:.0f}")
    print(f"Maximum Revenue: ${max_revenue:.2f}")
else:
    print("❌ No optimal solution found.")
```

### ✏️ Python Code Using scipy.optimize.linprog

```python
from scipy.optimize import linprog

# Coefficients for the objective function
c = [11, 6, 7]  # Minimize g = 11y1 + 6y2 + 7y3

# Inequality constraints (Ax <= b), convert ≥ to ≤
A = [
    [-2, -4, -5],
    [-3, -2, -4],
    [-5, -2, -3]
]
b = [-4, -3, -7]

# Bounds for y1, y2, y3
bounds = [(0, None), (0, None), (0, None)]

# Solve using HiGHS
res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Output results
if res.success:
    print("Optimal solution found:")
    print(f"y1 = {res.x[0]:.4f}")
    print(f"y2 = {res.x[1]:.4f}")
    print(f"y3 = {res.x[2]:.4f}")
    print(f"Minimum value of g = {res.fun:.4f}")
else:
    print("No optimal solution found.")
    print(res.message)
```

#### 6)For the following objective function and constraints: Maximize  P=3x+4y  subject to 7x+12y<=26 , 17x+20y<=40 , x,y>=0
#### Write the initial Simplex tableau:
#### Solution:-
#### Step 1: Identify the Pivot Column:-Look at the Z-row (last row) for the most negative coefficient. This will indicate the entering variable.
#### Z-row: [-3, -4, 0, 0] The most negative value is -4, which corresponds to y.Therefore, y is the entering variable.
#### Step 2: Identify the Pivot Row:-Now, we compute the ratio of the RHS to the coefficients in the pivot column (y) for each constraint. The smallest #### positive ratio will give us the leaving variable.
#### For s₁:26 12 =2.17 For s₂: 40 20=2   The smallest ratio is 2, corresponding to s₂ (second row). Therefore, s₂ is the leaving variable.
#### Step 3: Perform Pivoting  :- Now, we'll pivot on the element in the tableau where y and s₂ intersect, which is 20 in the second row and second column.
#### Pivoting Steps: Divide the pivot row (s₂) by the pivot element (20): ⋅[17,20,0,1,40]=[0.85,1,0,0.05,2]
#### Update the other rows by making the coefficient of y in all other rows zero.For this, we need to subtract multiples of the new s₂ row from the other #### rows.
#### Step 4: Update Tableau:-After performing the pivoting, we get a new tableau:
####    x	                 y	            s₁	  s₂	                     RHS	    P
#### s₁	7 - 7(0.85) = 2.05	12 - 12(1) = 0	1	0 - 0.05(1) = -0.05	26 - 2(2) = 22	    0 
#### s₂	17 / 20 = 0.85	20 / 20 = 1	0	1 / 20 = 0.05	40 / 20 = 2	                    0
#### Z	-3 + 3(0.85) = 0.55	-4 + 4(1) = 0	0	0 - 0.05(1) = -0.05	0 + 0.05(2) = 0.1	1
#### Now, we have a new tableau. Notice the update in the Z-row and RHS values.
#### Step 5: Check for Optimality:-The coefficients in the Z-row are all non-negative: [0.55, 0, 0, -0.05].Therefore, the current solution is optimal.
#### Final Solution:- After pivoting, the optimal values of the variables are:
#### Variable	Value
####  x	       2.05
####  y	        0
####  s₁	22
#### s₂   	2
#### P	   0.1
#### Explanation:The optimal value of the objective function (P) is 0.1.The final values of x and y are 2.05 and 0, respectively.s₁ and s₂ are the slack #### variables, with s₁ = 22 and s₂ = 2.

```python
from scipy.optimize import linprog

# Coefficients of the objective function (negated because linprog does minimization)
c = [-3, -4]

# Coefficients of the inequality constraints
A = [
    [7, 12],
    [17, 20]
]

# Right-hand side of the inequality constraints
b = [26, 40]

# Bounds for x and y (both >= 0)
bounds = [(0, None), (0, None)]

# Solve using the simplex method
res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Print results
print("Optimal value (Max P):", -res.fun)  # Negate because we minimized -P
print("Values of x and y:", res.x)
```

#### 8)Rutherford Coffee Shop sells two basic blends: Humberano and CrustMont.They use two types of coffee to produce the blends: BlueMont and Java.
#### Humberano sells for $6.50 per pound and requires at least 20% BlueMont. CrustMont sells for (dollar)5.50 per pound and requires no more than 30% Java.
#### The shop has 85 pounds of BlueMont and 110 pounds of Java. BlueMont costs (dollar)3.00 per pound while Java costs $4.00 per pound.
#### Let xij  represent pounds of coffee  i  in blend  j
#### The objective is to maximize profit.

```python
#solution:- #To derive the objective function for maximizing total profit, we need to account for the following details:
# Profit for Each Coffee Blend:The goal is to calculate the total profit from selling Humberano and CrustMont, considering the revenue from 
# selling the blends and the cost of the coffee beans used in those blends.
# 1. Profit from Humberano:Humberano sells for $6.50 per pound.
# It requires BlueMont and Java as ingredients. The amounts of BlueMont and Java are represented by the decision variables 
# - xBH  (BlueMont in Humberano) and xJH(Java in Humberano).
# To calculate the profit, we need to subtract the cost of the coffee used from the selling price:
# BlueMont costs $3.00 per pound.  Java costs $4.00 per pound.
# So, the profit per pound of Humberano will be: BlueMont contributes: 6.50−3.00=3.50 dollars per pound.
# Java contributes: 6.50−4.00=2.50 dollars per pound.
# Thus, the total profit from Humberano is: Profit Humberano =3.50⋅x BH  +2.50⋅x JH
# 2. Profit from CrustMont:CrustMont sells for $5.50 per pound.
# It also requires BlueMont and Java as ingredients. The amounts of BlueMont and Java are represented by the decision variables 
# 𝑥𝐵𝐶  (BlueMont in CrustMont) and xJC (Java in CrustMont).
# Similarly, we calculate the profit by subtracting the cost of the coffee beans from the selling price:
# BlueMont costs $3.00 per pound. Java costs $4.00 per pound.
# So, the profit per pound of CrustMont will be:BlueMont contributes: 5.50−3.00=2.50 dollars per pound.
# Java contributes: 5.50−4.00=1.50 dollars per pound.
#Thus, the total profit from CrustMont is:Profit CrustMont =2.50⋅x BC +1.50⋅xJC
# Total Profit:Now that we have the profit from both blends, the total profit is the sum of the individual profits:
#Total Profit=Profit Humberano+Profit CrustMont
# Substituting the expressions for the profits of each blend:Total Profit=(3.50⋅xBH​ +2.50⋅xJH )+(2.50⋅x BC +1.50⋅xJC )
# Simplifying:Total Profit=3.50⋅x BH​ +2.50⋅x JH​ +2.50⋅x BC +1.50⋅x JC​
```

```python
pip install pulp
```

```python
import pulp

# Create a linear programming problem (maximization)
prob = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Define decision variables (amount of BlueMont and Java in each blend)
x_BH = pulp.LpVariable("x_BH", lowBound=0, cat='Continuous')  # BlueMont in Humberano
x_JH = pulp.LpVariable("x_JH", lowBound=0, cat='Continuous')  # Java in Humberano
x_BC = pulp.LpVariable("x_BC", lowBound=0, cat='Continuous')  # BlueMont in CrustMont
x_JC = pulp.LpVariable("x_JC", lowBound=0, cat='Continuous')  # Java in CrustMont

# Objective function: maximize profit
prob += 3.50 * x_BH + 2.50 * x_JH + 2.50 * x_BC + 1.50 * x_JC, "Total_Profit"

# Constraints:

# BlueMont supply constraint (85 pounds available)
prob += x_BH + x_BC <= 85, "BlueMont_Supply"

# Java supply constraint (110 pounds available)
prob += x_JH + x_JC <= 110, "Java_Supply"

# Humberano blend constraint (at least 20% BlueMont)
prob += x_BH >= 0.2 * (x_BH + x_JH), "Humberano_BlueMont_Constraint"

# CrustMont blend constraint (no more than 30% Java)
prob += x_JC <= 0.3 * (x_BC + x_JC), "CrustMont_Java_Constraint"

# Solve the problem
prob.solve()

# Output the results
print(f"Status: {pulp.LpStatus[prob.status]}")
print(f"Optimal amount of BlueMont in Humberano (x_BH): {x_BH.varValue} pounds")
print(f"Optimal amount of Java in Humberano (x_JH): {x_JH.varValue} pounds")
print(f"Optimal amount of BlueMont in CrustMont (x_BC): {x_BC.varValue} pounds")
print(f"Optimal amount of Java in CrustMont (x_JC): {x_JC.varValue} pounds")

# Calculate and print the total amount of coffee used (sum of all variables)
total_coffee = x_BH.varValue + x_BC.varValue + x_JH.varValue + x_JC.varValue
print(f"Total amount of coffee used (BlueMont + Java in both blends): {total_coffee} pounds")

# Total Profit
print(f"Total Profit: ${pulp.value(prob.objective):.2f}")
```

Each newspaper ad reaches 5650 potential customers, and each  #### radio ad reaches 6950 potential customers. The company wants to maximize the number of ad exposures to potential customers.
#### Use n for number of Newspaper advertisements and r for number of Radio advertisements. Set up the linear programming model.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Define the objective function coefficients (negative for maximization)
c = [-5650, -6950]  # Maximize z = 5650n + 6950r

# Inequality constraints:
# 130n + 540r <= 14210
# n <= 18
# r <= 28
A = [
    [130, 540],
    [1, 0],
    [0, 1]
]
b = [14210, 18, 28]

# Variable bounds (n >= 0, r >= 0)
x_bounds = (0, None)
y_bounds = (0, None)

# Solve the LP problem
res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Extract and round results
n_opt = round(res.x[0])
r_opt = round(res.x[1])
max_exposure = round(-res.fun)

# Print result
print("Optimal Number of Newspaper Ads (n):", n_opt)
print("Optimal Number of Radio Ads (r):", r_opt)
print("Maximum Exposure (people):", max_exposure)

# Plot the feasible region
n = np.linspace(0, 30, 400)
r_budget = (14210 - 130 * n) / 540
r_max = np.full_like(n, 28)

plt.figure(figsize=(10, 6))
plt.plot(n, r_budget, label='130n + 540r ≤ 14210', color='blue')
plt.axvline(18, label='n ≤ 18', color='green', linestyle='--')
plt.axhline(28, label='r ≤ 28', color='orange', linestyle='--')
plt.fill_between(n, 0, np.minimum(r_budget, r_max), where=(n <= 18), color='lightgrey', alpha=0.5)
plt.scatter(n_opt, r_opt, color='red', zorder=5, label='Optimal Solution')
plt.xlim(0, 30)
plt.ylim(0, 35)
plt.xlabel('Newspaper Ads (n)')
plt.ylabel('Radio Ads (r)')
plt.title('Feasible Region and Optimal Advertising Plan')
plt.legend()
plt.grid(True)
plt.show()
```

#### 11)A factory manufactures three products, A, B, and C. Each product requires the use of two machines, Machine I and Machine II. The total hours   #### available, respectively, on Machine I and Machine II per month are 7,020 and 9,810. The time requirements and profit per unit for each product are #### listed below.
#### 	A	B	C
#### Machine I	5	10	8
#### Machine II	9	10	13
#### Profit	$9	$15	$21
#### How many units of each product should be manufactured to maximize profit, and what is the maximum profit?
#### Start by setting up the linear programming problem, with A, B, and C representing the number of units of each product that are produced.

```python
from scipy.optimize import linprog

# Coefficients of the objective function (maximize Z = 9A + 15B + 21C)
# We convert this to minimization by negating the coefficients
c = [-9, -15, -21]

# Coefficients for the inequality constraints (Machine I and Machine II)
A = [
    [5, 10, 8],   # Machine I hours constraint
    [9, 10, 13]   # Machine II hours constraint
]

# Right-hand side values of the constraints
b = [7020, 9810]

# Variable bounds: A, B, C >= 0
bounds = [(0, None), (0, None), (0, None)]

# Solve using HiGHS method
res = linprog(c=c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Display results
if res.success:
    # Round the results to 1 decimal place for units and 2 decimal places for profit
    A_units = round(res.x[0], 1)
    B_units = round(res.x[1], 1)
    C_units = round(res.x[2], 1)
    max_profit = round(-res.fun, 2)

    print("Optimal solution found:")
    print(f"A = {A_units} units")
    print(f"B = {B_units} units")
    print(f"C = {C_units} units")
    print(f"Maximum Profit = ${max_profit}")
else:
    print("No optimal solution found.")
```

#### 12)A lawn seed mixture contains three types of seed: bluegrass, rye, and Bermuda. The costs per pound of the three types of seed are 15 cents, 16 #### cents, and 4 cents, respectively. In each batch there must be at least 20% bluegrass seed, and the amount of Bermuda must be no more than 2/3 the #### amount of rye. To fill current orders, the company must make at least 8000 pounds of the mixture. How much of each kind of seed should be used to #### minimize cost? Round your answers to two decimals as needed.

```python
from scipy.optimize import linprog

# Coefficients of the objective function to minimize (costs per pound)
costs = [0.15, 0.16, 0.04]  # [bluegrass, rye, bermuda]

# Inequality constraints (Ax <= b)
# Constraint 1: -x - y - z <= -8000 (total weight >= 8000)
# Constraint 2: -0.8x + 0.2y + 0.2z <= 0 (bluegrass >= 20%)
# Constraint 3: 0y - 3z <= 0 (bermuda <= 2/3 rye => -2y + 3z <= 0, multiplied by -1)
A = [[-1, -1, -1],
     [-0.8, 0.2, 0.2],
     [0, -2, 3]]
b = [-8000, 0, 0]

# Bounds for each variable (non-negative amounts)
bounds = [(0, None), (0, None), (0, None)]  # for x, y, z respectively

# Solve the linear programming problem
result = linprog(costs, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Extract the optimal amounts and cost
optimal_amounts = result.x
minimum_cost = result.fun

# Print the results
if result.success:
    print("Optimal amounts of seed:")
    print(f"Bluegrass: {optimal_amounts[0]:.2f} pounds")
    print(f"Rye: {optimal_amounts[1]:.2f} pounds")
    print(f"Bermuda: {optimal_amounts[2]:.2f} pounds")
    print(f"Minimum cost: ${minimum_cost:.2f}")
else:
    print(f"Optimization failed: {result.message}")
```

#### 13) Julie designs magazine covers and brochures. Each magazine requires 6 hours of brainstorming and 2 hours of layout. Each brochure requires 3  #### hours of brainstorming and 2 hours of layout. The company makes a profit of (dolar)150 from each magazine cover and $110 from each brochure. She has #### a maximum of 39 hours for brainstorming before the deadline and 16 hours for layout before the deadline.
#### Let m represent the number of magazine covers to design and let b represent the number of brochures to design. Set up the linear programming model #### for maximizing profits.

```python
from scipy.optimize import linprog

# Coefficients of the objective function to maximize (negative of profits, as linprog minimizes)
# [magazine profit, brochure profit]
profits = [-150, -110]

# Inequality constraints (Ax <= b)
# Constraint 1: 6m + 3b <= 39 (brainstorming hours)
# Constraint 2: 2m + 2b <= 16 (layout hours)
A = [[6, 3],
     [2, 2]]
b = [39, 16]

# Bounds for each variable (non-negative number of designs)
# (min, max) - None implies no upper bound
bounds = [(0, None), (0, None)]  # for m (magazines) and b (brochures)

# Solve the linear programming problem
result = linprog(profits, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Extract the optimal number of magazines and brochures and the maximum profit
optimal_m = result.x[0]
optimal_b = result.x[1]
max_profit = -result.fun  # Negate the result because linprog minimizes

# Print the results
if result.success:
    print("Optimal number of designs:")
    print(f"Magazine covers (m): {optimal_m:.2f}")
    print(f"Brochures (b): {optimal_b:.2f}")
    print(f"Maximum profit: ${max_profit:.2f}")
else:
    print(f"Optimization failed: {result.message}")
```

#### 14) The water-supply manager for New York City needs to supply the city with at least 17 million gallons of potable water per day. The supply may be #### drawn from the local reservoir or from a pipeline to an adjacent town. The local reservoir has a maximum daily yield of 18 million gallons of potable #### water, and the pipeline has a maximum daily yield of 12 million gallons. By contract, the pipeline is required to supply a minimum of 7 million   #### gallons per day. Suppose the cost for 1 million gallons of reservoir water is (dolar)485 and the cost for 1 million gallons of pipeline water is $415.
#### Let p be the number of gallons of water (in millions) drawn from the pipeline and let r be the number of gallons of water (in millions) drawn from #### the reservoir. State the objective function for minimizing costs.


```python
from scipy.optimize import linprog

# Coefficients of the objective function to minimize (costs per million gallons)
# [pipeline cost, reservoir cost]
costs = [415, 485]

# Inequality constraints (Ax <= b)
# Constraint 1: -p - r <= -17 (total supply >= 17 million gallons)
# Constraint 2: 0p + r <= 18 (reservoir yield <= 18 million gallons)
# Constraint 3: p + 0r <= 12 (pipeline yield <= 12 million gallons)
# Constraint 4: -p + 0r <= -7 (pipeline supply >= 7 million gallons)
A = [[-1, -1],
     [0, 1],
     [1, 0],
     [-1, 0]]
b = [-17, 18, 12, -7]

# Bounds for each variable (non-negative amounts and pipeline minimum)
# (min, max) - None implies no upper bound
bounds_p = (7, 12)  # Pipeline: min 7, max 12
bounds_r = (0, 18)  # Reservoir: min 0, max 18
bounds = [bounds_p, bounds_r]

# Solve the linear programming problem
result = linprog(costs, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Extract the optimal amounts and minimum cost
optimal_pipeline = result.x[0]
optimal_reservoir = result.x[1]
min_cost = result.fun

# Print the results
if result.success:
    print("Optimal water supply (in millions of gallons):")
    print(f"Pipeline (p): {optimal_pipeline:.2f}")
    print(f"Reservoir (r): {optimal_reservoir:.2f}")
    print(f"Minimum cost: ${min_cost:.2f}")
else:
    print(f"Optimization failed: {result.message}")
```
