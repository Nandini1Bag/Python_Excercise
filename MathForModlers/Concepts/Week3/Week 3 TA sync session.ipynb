### 2. Use Sympy library package

```python
#Here we convert inequalities to equalities to find the nodes using Sympy library package

import sympy as sym
x,y = sym.symbols('x, y')

#Let us define the equations to represent inequalities
eq1 = x + 3*y -10
eq2 = 3*x + y -6
eq3 = y
eq4 = x

print('Our equations are:')
print('Eq1:',eq1,'\nEq2:',eq2,'\nEq3:',eq3,'\nEq4:',eq4)


#Now, let us find the simplex nodes
print('\nThe roots of every pair of equations are:')
print('Eq1 & Eq2:',sym.solve([eq1,eq2],[x,y]))
print('Eq1 & Eq4:',sym.solve([eq1,eq4],[x,y]))
print('Eq2 & Eq3:',sym.solve([eq2,eq3],[x,y]))
print('Eq3 & Eq4:',sym.solve([eq3,eq4],[x,y]))

#Let us define the objective functions
Obj = lambda x,y: 5*x + 2*y

#Evaluating the objective function at the root
print("\nSolution (x:1, y:3) = ", Obj(1,3))
print("Solution (x:0, y:10/3) = ", Obj(0,10/3))
print("Solution (x:2, y:0) = ", Obj(2,0))
print("Solution (x:0, y:0) = ", Obj(0,0))
```

### 3. Use Scipy library method - linprog to find solution

```python
#First of all, we will use the Scientific python library package
#Ref: https://realpython.com/linear-programming-python/
#Ref: https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.linprog.html

from scipy.optimize import linprog

#Let us define the3 matrices
obj = np.array([-5,-2])            #The objective function
lhs = np.array([[1,3],[3,1]])      #The left hand side of the equations
rhs = np.array([10,6])             #The right hand side of the equations

bnd = [(0,float('inf')),(0,float('inf'))]   #This is boundary of our decision variables used

results = linprog(c = obj, A_ub = lhs, b_ub = rhs, bounds = bnd) #, method = 'highs'

if results.status == 0: print('The solution is optimal.') 
print(f'Solution:',results.x)
print(f'Objective value: z = {-results.fun}')
```

### 4. Use Pulp

```python

# Import all classes of PuLP module
from pulp import LpProblem, LpVariable, LpMaximize, LpInteger, value

# Create the problem variable to contain the problem data
model = LpProblem("LinearProgProblem", LpMaximize)

# Create variables
x = LpVariable("x", 0, None, LpInteger)
y = LpVariable("y", 0, None, LpInteger) 

# Create maximize objective function
model += 5*x + 2*y 

# Create constraints
model +=  x + 3*y   <= 10,  "C1"
model +=  3*x + y   <= 6, "C2"

# The problem is solved using PuLP's choice of Solver
model.solve()

# Each of the variables is printed with it's resolved optimum value
for v in model.variables():
    print(v.name, "=", v.varValue)
    
print("Value of the objective function ",model.objective,"is :",value(model.objective))
```

### 5. Google OR tools

```python
from ortools.linear_solver import pywraplp
#Ref: https://medium.com/@chongjingting/4-ways-to-solve-linear-programming-in-python-b4af36b7894d
#Ref: https://developers.google.com/optimization/reference/python/linear_solver/pywraplp

# get solver 
solver = pywraplp.Solver.CreateSolver('GLOP') #Google Linear Optimization Package. Other solvera: CBC, CLP etc.
#Ref:https://or-tools.github.io/docs/pdoc/ortools/linear_solver/pywraplp.html

# declare decision variables
x = solver.NumVar(0.0, solver.infinity(), 'x')
y = solver.NumVar(0.0, solver.infinity(), 'y')

# declare objective
solver.Maximize(5*x + 2*y )

# declare constraints
solver.Add(x + 3*y <= 10)
solver.Add(3*x + y <= 6)


# solve
results = solver.Solve()

# print results
if results == pywraplp.Solver.OPTIMAL: print(f'The solution is optimal.') 
 
print(f'Solution: x = {np.round(x.solution_value(),3)}, y = {np.round(y.solution_value(),3)}')
print(f'Objective value: z = {np.round(solver.Objective().Value(),3)}')
```
