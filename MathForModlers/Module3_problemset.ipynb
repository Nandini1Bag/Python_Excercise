#### 1)graph x-y>=-3 , 2x+6y<=12  

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(-10, 10, 400)

# Boundary lines
y1 = x + 3                    # from x - y ≥ -3 → y ≤ x + 3
y2 = (-1/3) * x + 2           # from 2x + 6y ≤ 12 → y ≤ -1/3x + 2

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot the boundary lines
plt.plot(x, y1, label='y = x + 3', color='blue')
plt.plot(x, y2, label='y = -1/3x + 2', color='green')

# Fill the feasible region
# Create a grid of points
X, Y = np.meshgrid(x, np.linspace(-10, 10, 400))

# Define the inequalities
ineq1 = Y <= X + 3
ineq2 = Y <= (-1/3) * X + 2

# Region that satisfies both
region = ineq1 & ineq2

# Plot the region
plt.contourf(X, Y, region, levels=[0.5, 1], colors=['lightgray'])

# Axis and grid
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the system of inequalities')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2)Caitlyn sells cell phone sweaters and keychains at the county fair. She wants to sell at least 10 crafts in total. She sells the cell phone 
#### sweaters for (dollar)2.00 and the keychains for (dollar)1.00. She needs to sell at least $13 worth of crafts in order to earn a profit.
#### a) Choose the system of inequalities that matches the given problem (where  is the number of cell phone sweaters and  is the number of keychains).
#### b) Will she make a profit if she sells 3 cell phone sweaters and 4 keychains?

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(0, 15, 400)

# Define the boundary lines
y1 = -x + 10       # from x + y >= 10
y2 = -2*x + 13     # from 2x + y >= 13

# Create a grid of points
X, Y = np.meshgrid(x, np.linspace(0, 15, 400))

# Inequality conditions
ineq1 = X + Y >= 10
ineq2 = 2*X + Y >= 13

# Feasible region (where both inequalities are true)
region = ineq1 & ineq2

# Plot the boundary lines
plt.plot(x, y1, label='x + y = 10', color='blue')
plt.plot(x, y2, label='2x + y = 13', color='green')

# Fill the feasible region
plt.contourf(X, Y, region, levels=[0.5, 1], colors=['lightgray'])

# Customize the plot
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.xlabel('x (Cell phone sweaters)')
plt.ylabel('y (Keychains)')
plt.title('System of Inequalities: Sales at the Fair')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

#### 3)Miranda runs a restaurant that sells two kinds of desserts. Miranda knows the restaurant must make at least 8 and at most 28 batches of the Mint #### Breezes. The restaurant must also make at least 17 and at most 60 batches of the Caramel Twists. Each batch of Mint Breezes takes 5 minutes in the #### oven, while each batch of Caramel Twists requires 4 minutes in the oven. The restaurant only has 320 minutes in the oven available. Suppose each   #### batch of Mint Breezes generates (dollar)2.91 in revenue, each batch of Caramel Twists generates (dollar)1.31, and Miranda needs to maximize revenue.
#### Let x be the number of batches of Mint Breezes and let y be the number of batches of Caramel Twists. Set up the linear programming model.

```python
from scipy.optimize import linprog

# Objective function coefficients (negated for maximization)
# Revenue: Z = 2.91x + 1.31y ⇒ maximize Z
# We negate because linprog minimizes by default
c = [-2.91, -1.31]

# Inequality constraint matrix (for oven time): 5x + 4y ≤ 320
A = [[5, 4]]
b = [320]

# Variable bounds
# x (Mint Breezes): between 8 and 28
# y (Caramel Twists): between 17 and 60
x_bounds = (8, 28)
y_bounds = (17, 60)

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Check if the optimization was successful
if result.success:
    # Extract values
    x_opt = result.x[0]
    y_opt = result.x[1]
    max_revenue = -result.fun  # negate back to get maximum

    print("✅ Optimal Solution Found:")
    print(f"Mint Breezes (x): {x_opt:.0f}")
    print(f"Caramel Twists (y): {y_opt:.0f}")
    print(f"Maximum Revenue: ${max_revenue:.2f}")
else:
    print("❌ No optimal solution found.")
```

#### 4)Use the information below to create the initial simplex tableau.Maximize P =19x1 + 5x2   subject to 28x1 23x2<=55 
####  15x1+ 17x2<=70  x1>=0 and x2>=0 



```python
import numpy as np
import pandas as pd

# Coefficients for the tableau
# Rows: constraints + objective function
# Columns: x1, x2, s1, s2, RHS

tableau = np.array([
    [28, 23, 1, 0, 55],   # Constraint 1
    [15, 17, 0, 1, 70],   # Constraint 2
    [-19, -5, 0, 0, 0]    # Objective function (Z)
])

# Create a pandas DataFrame for better readability
columns = ['x1', 'x2', 's1', 's2', 'RHS']
index = ['Constraint 1 (s1)', 'Constraint 2 (s2)', 'Z']

tableau_df = pd.DataFrame(tableau, columns=columns, index=index)

print("Initial Simplex Tableau:")
print(tableau_df)
```

#### 5)Create the initial simplex tableau.
#### Minimize g = 11y1+ 6y2 +7y3 subject to  2y1+4y2+5y3>=4 , 3y1+ 2y2+4y3>=3 , 5y1+2y2+3y3>=7
#### y1>=0,y2>=0,y3>=0


```python
import numpy as np
import pandas as pd

# Big M value (symbolic representation often used in theory; use a large number in practice)
M = 1e6

# Coefficients for the initial tableau (variables: y1, y2, y3, s1, s2, s3, a1, a2, a3, RHS)
tableau = np.array([
    [2,  4, 5, -1,  0,  0, 1, 0, 0,  4],  # Constraint 1
    [3,  2, 4,  0, -1,  0, 0, 1, 0,  3],  # Constraint 2
    [5,  2, 3,  0,  0, -1, 0, 0, 1,  7],  # Constraint 3
    [-11, -6, -7, 0, 0, 0, -M, -M, -M, 0] # Objective function (Big M method)
])

# Define the column names and index
columns = ['y1', 'y2', 'y3', 's1', 's2', 's3', 'a1', 'a2', 'a3', 'RHS']
index = ['Constraint 1', 'Constraint 2', 'Constraint 3', 'G (Objective)']

# Create the DataFrame to represent the tableau
df = pd.DataFrame(tableau, columns=columns, index=index)

# Print the initial simplex tableau
print("Initial Simplex Tableau:")
print(df)
```
