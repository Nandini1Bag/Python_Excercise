#### 1)Solve the system by graphing y=x+9   y=-5/6x-2

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values
x = np.linspace(-10, 5, 400)

# Define the two functions
y1 = x + 9
y2 = -5/6 * x - 2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = x + 9', color='blue')
plt.plot(x, y2, label='y = -5/6x - 2', color='green')

# Plot the intersection point
plt.plot(-6, 3, 'ro')  # red point at (-6, 3)
plt.text(-6, 3.5, '(-6, 3)', fontsize=12, color='red')

# Add labels and grid
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphical Solution of the System')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

#### 2)Treyvon has a total of 186 vitamin supplement tablets comprised of three types of tablets. Each tablet A  contains 12g mg of niacin and  10 g I.U. ####  of vitamin E. Each tablet  B contains 17g mg of niacin and  17g I.U of vitamin E. Each tablet  C contains 22g  mg of niacin and  33g I.U. of vitamin ####  E. Suppose the entire mixture of  186 tablets contains 3412 mg of niacin and  4295 I.U. of vitamin E. 





```python
import numpy as np

# Coefficient matrix A
A = np.array([
    [1, 1, 1],
    [12, 17, 22],
    [10, 17, 33]
])

# Constants matrix B
B = np.array([186, 3412, 4295])

# Solve for X
X = np.linalg.solve(A, B)

# Print the result
a, b, c = X
print(f"Tablet A: {int(a)}, Tablet B: {int(b)}, Tablet C: {int(c)}")
```

#### 5)Find the eigenvalues of A  and A^-1 given the following matrix: [[ 1 7 -3 ][0 4 2  ][ 0 0 2]]

```python
import numpy as np

# Define the matrix A
A = np.array([
    [1, 7, -3],
    [0, 4, 2],
    [0, 0, 2]
])

# Eigenvalues of A
eigenvalues_A = np.linalg.eigvals(A)

# Inverse of A
A_inv = np.linalg.inv(A)

# Eigenvalues of A inverse
eigenvalues_A_inv = np.linalg.eigvals(A_inv)

# Print results
print("Eigenvalues of A:", eigenvalues_A)
print("Eigenvalues of A⁻¹:", eigenvalues_A_inv)
```

```python
import numpy as np
def is_upper_triangular(matrix):
    return np.allclose(matrix, np.triu(matrix))

def is_lower_triangular(matrix):
    return np.allclose(matrix, np.tril(matrix))
    
# np.triu(matrix) → gets the upper triangular part of the matrix.
# np.tril(matrix) → gets the lower triangular part.
# np.allclose() checks if two arrays are approximately equal (handles tiny rounding errors).

# Example matrix (you can change this)
A = np.array([
    [1, 7, -3],
    [0, 4, 2],
    [0, 0, 2]
])

# Check
if is_upper_triangular(A):
    print("Matrix is upper triangular.")
elif is_lower_triangular(A):
    print("Matrix is lower triangular.")
else:
    print("Matrix is not triangular.")
```
